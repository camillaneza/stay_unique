```python
# Agrupar por PropertyId y ArrivalDate y contar las ocurrencias
duplicates = df2.groupby(['PropertyId', 'ArrivalDate']).size().reset_index(name='Count')

# Filtrar solo los duplicados (Count > 1)
duplicate_entries = duplicates[duplicates['Count'] > 1]

# Mostrar los duplicados
print("Duplicados encontrados:")
print(duplicate_entries)
```

```python
# Si deseas ver todas las filas del DataFrame original que corresponden a estos duplicados
if not duplicate_entries.empty:
    # Obtener las combinaciones de PropertyId y ArrivalDate duplicados
    duplicated_ids = duplicate_entries[['PropertyId', 'ArrivalDate']]

    # Filtrar el DataFrame original para que solo contenga las coincidencias
    df_duplicates = df2.merge(duplicated_ids, on=['PropertyId', 'ArrivalDate'])

    # Mostrar el DataFrame original con las coincidencias
    print("\nFilas duplicadas en el DataFrame original:")
    print(df_duplicates)
else:
    print("No hay duplicados encontrados.")
```

```python
duplicates = df2.groupby(['PropertyId', 'ArrivalDate']).size().reset_index(name='Count')

# Filtrar solo los duplicados (Count > 1)
duplicate_entries = duplicates[duplicates['Count'] > 1]

# Unir con el DataFrame original para obtener más información, incluyendo Property_BookingId
merged_duplicates = duplicate_entries.merge(df2, on=['PropertyId', 'ArrivalDate'], how='inner')

# Mostrar los duplicados y verificar coincidencias en Property_BookingId
print("Duplicados encontrados:")
merged_duplicates[['PropertyId', 'ArrivalDate', 'Property_BookingId', 'Count']].head(30)
```

```python
# Agrupar por PropertyId y ArrivalDate y contar las ocurrencias
duplicates = df2.groupby(['PropertyId', 'ArrivalDate']).size().reset_index(name='Count')

# Filtrar solo los duplicados (Count > 1)
duplicate_entries = duplicates[duplicates['Count'] > 1]

# Para cada entrada duplicada, obtener las fechas de salida (DepartureDate)
for index, row in duplicate_entries.iterrows():
    property_id = row['PropertyId']
    arrival_date = row['ArrivalDate']
    
    # Filtrar el DataFrame original para ver las fechas de salida correspondientes
    matched_rows = df2[(df2['PropertyId'] == property_id) & (df2['ArrivalDate'] == arrival_date)]
    
    print(f"PropertyId: {property_id}, ArrivalDate: {arrival_date}")
    print(matched_rows[['PropertyId', 'ArrivalDate', 'DepartureDate']])
    print("\n")
```
